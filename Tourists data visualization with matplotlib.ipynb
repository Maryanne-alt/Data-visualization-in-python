```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tourist_data=pd.read_csv('Train.csv')
```

```python
tourist_data.head()
```

```python
tourist_data.shape
```

```python
tourist_data.describe()
```

```python
tourist_data.country.value_counts()
```

```python
tourist_data.country.value_counts().plot.bar(figsize=(25,5))
plt.title('distribution of frequency countries')
plt.ylabel('frequency')
```

```python
tourist_data.country.value_counts(normalize=True)*100
```

```python
(tourist_data.country.value_counts(normalize=True)*100).plot.bar(figsize=(25,5))
plt.title('percentage disribution of countries')
plt.ylabel('percentage')
```

```python
(tourist_data.travel_with.value_counts(normalize=True)*100).plot.bar(figsize=(10,5))
plt.title('percentage disribution of travel with')
plt.ylabel('percentage')
plt.xlabel('travel with')
```
