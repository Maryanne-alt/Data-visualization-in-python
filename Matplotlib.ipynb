```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Gender = np.array(['male','female','female','male','male','female','male','female','female','female','male'])
Mathematics = np.array(['78','72','55','67','70','75','69','66','63','58','61'])
plt.plot(Gender,Mathematics)
plt.title('Gender vs Mathematics')
plt.xlabel('Gender')
plt.ylabel('Mathematics')
```

```python
Gender = ['male','female','female','male','male','female','male','female','female','female','male']
Mathematics = ['78','72','55','67','70','75','69','66','63','58','61']
English= ['70','79','75','69','62','75','65','68','65','63','58']
plt.plot(Gender,Mathematics, color='red', marker='*')
plt.plot(Gender,English, color ='green', marker = 'o')
plt.title('Gender, Mathematics vs English')
plt.xlabel('Gender')
plt.ylabel('Mathematics and English')
plt.show()
           
```
